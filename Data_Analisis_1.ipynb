```python
# prompt: Filter the data based on a condition and update the plot to reflect the filtered data. by taking age on x axis and smoker on y axis and print the age of smoker who smokes more

import matplotlib.pyplot as plt

# Filter data for smokers
smokers_df = df[df['smoker'] == 'yes']

# Create the scatter plot with filtered data
plt.figure(figsize=(10, 6))
plt.scatter(smokers_df['age'], smokers_df['smoker'], alpha=0.7)
plt.xlabel('Age')
plt.ylabel('Smoker')
plt.title('Scatter Plot of Age vs. Smoker (Smokers Only)')
plt.grid(True)

# Find and print the age of the smoker with the highest charges
smoker_with_highest_charges = smokers_df.loc[smokers_df['charges'].idxmax()]
print(f"The age of the smoker with the highest charges is: {smoker_with_highest_charges['age']}")

plt.show()
```

```python
# prompt: if age IS  incresing whether  bmi , increasing or decresing or stable and print thaT  BY ANALIZING ABOVE CODES Print the graph and print age and incresing bmi

import pandas as pd
import matplotlib.pyplot as plt

# Assuming 'df' is your DataFrame (loaded as in your previous code)

# Group data by age and get the mean BMI for each age group
age_bmi_grouped = df.groupby('age')['bmi'].mean()

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(age_bmi_grouped.index, age_bmi_grouped.values, marker='o', linestyle='-')
plt.xlabel('Age')
plt.ylabel('Average BMI')
plt.title('Average BMI vs. Age')
plt.grid(True)

# Analyze and print BMI trend for increasing age
for i in range(1, len(age_bmi_grouped)):
    if age_bmi_grouped.iloc[i] > age_bmi_grouped.iloc[i-1]:
        print(f"Age: {age_bmi_grouped.index[i]}, BMI: {age_bmi_grouped.iloc[i]}, Trend: Increasing")
    elif age_bmi_grouped.iloc[i] < age_bmi_grouped.iloc[i-1]:
      print(f"Age: {age_bmi_grouped.index[i]}, BMI: {age_bmi_grouped.iloc[i]}, Trend: Decreasing")
    else:
      print(f"Age: {age_bmi_grouped.index[i]}, BMI: {age_bmi_grouped.iloc[i]}, Trend: Stable")

plt.show()
```

```python
# prompt: plot a statter plot with x=ages btw 20 to 40 age group vs smokes in that ages in red colout  and also give count '

import matplotlib.pyplot as plt

# Assuming 'df' is your DataFrame (loaded as in your previous code)

# Filter for the age group 20-40
age_group_df = df[(df['age'] >= 20) & (df['age'] <= 40)]

# Count the number of smokers in the age group
smoker_counts = age_group_df['smoker'].value_counts()

# Create the scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(age_group_df['age'], age_group_df['smoker'], color='red')
plt.xlabel('Age (20-40)')
plt.ylabel('Smoker (yes/no)')
plt.title('Smokers vs. Age (20-40)')
plt.grid(True)


# Annotate with smoker counts
for i, count in enumerate(smoker_counts):
    plt.annotate(f'{count}', (20, i), textcoords="offset points", xytext=(5, 0), ha='left')

plt.show()
```
